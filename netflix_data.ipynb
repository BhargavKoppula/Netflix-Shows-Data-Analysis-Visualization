# Data Cleaning and Visualization of Netflix_tiitles dataset using pandas, matplotliob and seaborn

```python
import pandas as pd

df = pd.read_csv('netflix_titles.csv')
```

### imported pandas and loaded it into df variable
and checking the data

```python
df.head()
```

```python
df.info()
```

### filling all the null values
 unknown in coloumns : director, cast, title, country, duration.
 
 and for rating mode is applied and the most occuring instance is placed

```python
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['title'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['duration'].fillna('Unknown', inplace=True)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)
```

```python
df
```

```python
df.dropna(subset=['title', 'director', 'country', 'cast'], inplace=True)
```

```python
df.isnull().sum()
```

changing the data format to datetime

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df.info()
```

### filling the null values in date with median_date

```python
median_date = df['date_added'].median()
median_date
```

```python
df['date_added'].fillna(median_date, inplace = True)
```

```python
df['duration_type'] = df['duration'].apply(lambda x: 'Season' if 'Season' in x else 'Minutes')
```

```python
df.isnull().sum()
```

```python
df['duration_num'] = df['duration'].str.extract('(\d+)').astype(float)
```

### removed all the null values in the dataset and them copying it to another dataset called netflix_final_data

```python
netflix_final_data = df.copy()
```

```python
netflix_final_data
```

```python
netflix_final_data.to_csv('netflix_titles_stage.csv', index =False)
```

```python
df.info()
```

```python
df = df[['show_id', 'type','title','director','cast','country','date_added','release_year','rating','duration','duration_num','duration_type','listed_in','description','genre_1','genre_2','genre_3']]
```

```python
df.tail()
```

```python
df[['genre_1', 'genre_2', 'genre_3']] = df['listed_in'].str.split(',', expand=True)
```

```python
netflix_final_data =df.copy()
```

```python
netflix_final_data.to_csv('netflix_titles_stage.csv')
```

```python
netflix_final_data.head()
```

### Distribution of Movies & TV Shows on Netflix

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count Movies & TV Shows
sns.countplot(data=netflix_final_data, x="type", palette="pastel")
plt.title("Distribution of Movies & TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

### Top 10 Most Popular Genres on Netflix

```python
plt.figure(figsize=(12,6))
sns.barplot(y=netflix_final_data['genre_1'].value_counts().head(10).index, 
            x=netflix_final_data['genre_1'].value_counts().head(10), 
            palette="coolwarm")

plt.title("Top 10 Most Popular Genres on Netflix")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```

### Netflix Content Growth Over the Years

```python
plt.Figure(figsize=(12,6))

netflix_final_data['year_added'] = netflix_final_data['date_added'].dt.year

sns.histplot(data=netflix_final_data, x="year_added", hue="type", multiple="stack", bins=20, palette="viridis")

plt.title("Content Added to Netflix Over the Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

### Average Movie Duration by Genre

```python
plt.figure(figsize=(12,6))

netflix_movies = netflix_final_data[netflix_final_data['type'] == 'Movie']

avg_duration = netflix_movies.groupby("genre_1")['duration_num'].mean().sort_values(ascending=False).head(10)

sns.barplot(y=avg_duration.index, x=avg_duration.values, palette="magma")

plt.title("Top 10 Genres with the Longest Average Movie Duration")
plt.xlabel("Average Duration (Minutes)")
plt.ylabel("Genre")
plt.show()

```

###  Distribution of Content Ratings

```python
plt.figure(figsize=(12,6))

sns.barplot(y=netflix_final_data['rating'].value_counts().index, 
            x=netflix_final_data['rating'].value_counts().values, 
            palette="coolwarm")

plt.title("Distribution of Content Ratings on Netflix")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.show()
```

### Top 10 Countries Producing Netflix Content

```python
plt.figure(figsize=(12,6))

# Count of titles by country (top 10)
sns.barplot(y=netflix_final_data['country'].value_counts().head(10).index, 
            x=netflix_final_data['country'].value_counts().head(10), 
            palette="plasma")

plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```
